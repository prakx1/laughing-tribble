```python
%matplotlib notebook
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
train_data=pd.read_csv('train.csv')
# train_data2=train_data1[pd.notnull(train_data1.Age)]
# train_data2
train_data['Age'].replace(np.nan,train_data['Age'].mean(),inplace=True)
```

```python
test_data=pd.read_csv('test.csv')
# test_data2=test_data1[pd.notnull(test_data1['Age'])]
test_data['Age'].replace(np.nan,test_data['Age'].mean(),inplace=True)
```

```python
train_data.head()
y=train_data['Survived']
features=['Pclass','Sex','Parch','SibSp','Age']
X=pd.get_dummies(train_data[features])
test_new=pd.get_dummies(test_data[features])
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(X,y)
knn.score(X,y)
```

```python
# scores=[]
# for i in range(1,30):
#     knn=KNeighborsClassifier(n_neighbors=2)
#     knn.fit(X,y)
#     knn.score(X,y)
#     scores.append((i,knn.score(X,y)))
# scores
preds=knn.predict(test_new)
preds
```

```python

output=pd.DataFrame({'PassengerId':test_data.PassengerId,'Survived':preds})
output['Survived'].value_counts(normalize=True)
```

```python
output.to_csv('knn3_age_mean.csv',index=False)
```
